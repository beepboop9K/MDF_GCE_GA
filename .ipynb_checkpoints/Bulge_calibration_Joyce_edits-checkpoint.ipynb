## Create function to make inflow rates that aren't in parallel

```python
def two_inflow_fn(t, exp_inflow):
    if t < exp_inflow[1][1]:
        inflow_rate = (exp_inflow[0][0]*np.exp(-t/exp_inflow[0][2]))
    else:
        inflow_rate = (exp_inflow[0][0]*np.exp(-t/exp_inflow[0][2])+
                           exp_inflow[1][0]*np.exp(-(t-exp_inflow[1][1])/exp_inflow[1][2]))
    return inflow_rate
```
